Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have?

```python
# Ans 1:-
import pandas as pd

flight_price_data = pd.read_excel("flight_price.xlsx")

num_rows, num_columns = flight_price_data.shape

print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
# Ans 2:-
import pandas as pd
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.hist(flight_price_data["Price"], bins=20, color="blue", edgecolor="black")
plt.title("Distribution of Flight Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
# Ans 3:-
import pandas as pd


min_price = flight_price_data["Price"].min()
max_price = flight_price_data["Price"].max()
price_range = max_price - min_price

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
# Ans 11:-
import pandas as pd

missing_values = data.isnull().sum()

print(missing_values)

missing_percentage = (missing_values / len(data)) * 100

print(missing_percentage)

high_missing_columns = missing_percentage[missing_percentage > 0]

for column, percentage in high_missing_columns.iteritems():
    print(f"Column: {column}, Missing Percentage: {percentage:.2f}%")
```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
# Ans 12:-
import pandas as pd
import matplotlib.pyplot as plt

filtered_data = data.dropna(subset=['Size', 'Rating'])

def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', '').replace(',', ''))
    if 'k' in size:
        return float(size.replace('k', '').replace(',', '')) / 1000
    return size

plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['Size'], filtered_data['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (MB)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
# Ans 14:-
import pandas as pd

filtered_data = data.dropna(subset=['Installs'])

def clean_installs(installs):
    return int(installs.replace(',', '').replace('+', '').replace('Free', '0'))

filtered_data['Installs'] = filtered_data['Installs'].apply(clean_installs)

sorted_data = filtered_data.sort_values(by='Installs', ascending=False)

top_apps = sorted_data.head(10)

print(top_apps[['App', 'Installs']])
```

Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the
Google Playstore dataset to identify the most popular app categories. How would you approach this
task, and what features would you analyze to make recommendations to the company?

```python
Ans 15:-
Data Preparation and Exploration:

Load the Google Playstore dataset and examine its structure.
Check for missing values in relevant columns.
Convert relevant columns to appropriate data types (e.g., 'Installs' to numeric).
Data Analysis:

Calculate the average number of installs for each app category.
Rank the app categories based on the average number of installs.
Visualization:

Create a bar chart or a horizontal bar chart to visualize the average number of installs for each app category.
Highlight the top app categories with the highest average installs.
Recommendations:

Identify the top app categories with the highest average installs.
Provide insights to the company about the most popular app categories based on user installs.
Suggest potential opportunities for the company to develop a new app in a popular category.
Further Analysis (Optional):

Analyze user ratings and reviews for the top app categories to ensure a balance between popularity and user satisfaction.
Consider analyzing the competition within the top app categories to assess the feasibility of entering those markets.
```

```python
Key Features to Analyze:

Category: This is the primary feature that determines the app category.
Installs: The number of installs indicates the popularity and demand for each app category.
Approach:

Load and prepare the dataset.
Calculate the average installs for each app category.
Rank the categories based on average installs.
Visualize the results using a bar chart.
Provide recommendations based on the findings.
```

```python
# Ans 15:-
import pandas as pd
import matplotlib.pyplot as plt

data['Installs'] = data['Installs'].str.replace(',', '').str.replace('+', '')

data = data[data['Installs'] != 'Free'] 
data['Installs'] = data['Installs'].astype(int)

avg_installs_by_category = data.groupby('Category')['Installs'].mean().sort_values(ascending=False)

top_categories = avg_installs_by_category.head(10)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular App Categories')
plt.xlabel('Category')
plt.ylabel('Average Installs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()

print("Recommendations for the company:")
for category, avg_installs in top_categories.items():
    print(f"{category}: Average Installs - {avg_installs:.0f}")
```
