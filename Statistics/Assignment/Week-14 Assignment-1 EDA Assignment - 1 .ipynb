Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
predicting the quality of wine.

```python
Ans 1:-
    The Wine Quality dataset contains various features that provide information about the chemical composition of different wines.
    These features play a crucial role in predicting the quality of wine.
```

```python
Fixed Acidity:
    Fixed acidity refers to the amount of non-volatile acids in the wine.
    It is an important factor in determining the wines overall taste and acidity level.
    Wines with balanced acidity are generally considered of higher quality.

Volatile Acidity:
    Volatile acidity represents the presence of volatile acids, primarily acetic acid, in the wine.
    Too much volatile acidity can result in a vinegar-like taste, which is undesirable in quality wines.

Citric Acid:
    Citric acid is a natural acid found in many fruits.
    It can contribute to the wines freshness and flavor balance.
    Adequate levels of citric acid can enhance the wine's overall appeal.

Residual Sugar:
    Residual sugar refers to the amount of sugar remaining in the wine after fermentation.
    It can impact the wines sweetness and mouthfeel.
    Balancing residual sugar is important for achieving the desired taste profile.

Chlorides:
    Chlorides can influence the wines taste and overall balance.
    Excessive chloride levels can lead to a salty or briny taste, affecting the wine's quality.

Free Sulfur Dioxide:
    Free sulfur dioxide acts as an antioxidant and preservative in wine.
    It helps prevent oxidation and microbial growth, which can impact the wine's longevity and quality.
```

Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.

```python
Ans 2:-
    Handling missing data is a crucial step in the feature engineering process to ensure accurate and reliable analysis.
    In the wine quality dataset, various techniques can be used to handle missing data. 
```

```python
Deletion of Missing Data (Listwise or Pairwise Deletion):
    Advantages:
        Simple and straightforward.
        Can be effective when the missing data is randomly distributed.
        Does not require imputing values.
    Disadvantages:
        Reduces the sample size, potentially leading to loss of valuable information.
        May introduce bias if the missing data is not completely random.
        
Mean, Median, or Mode Imputation:
    Advantages:
        Easy to implement and does not distort the distribution of the variable.
        Can work well for variables with a small proportion of missing values.
    Disadvantages:
        Ignores the relationships between variables.
        Can result in an underestimation of the variance.
        May not be suitable for variables with a large number of missing values.
        
Regression Imputation:
    Advantages:
        Takes into account the relationships between variables by using regression models.
        Provides more accurate imputed values compared to mean or median imputation.
    Disadvantages:
        Assumes a linear relationship between variables, which may not always hold true.
        Sensitive to outliers and multicollinearity.
        
K-Nearest Neighbors (KNN) Imputation:
    Advantages:
        Considers the similarity between samples to impute missing values.
        Can handle both numerical and categorical data.
        Can capture complex relationships.
    Disadvantages:
        Computationally intensive, especially for large datasets.
        Choice of k (number of neighbors) can impact imputation accuracy.
        May not work well with high-dimensional data.
```

Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution
of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to
these features to improve normality?

```python
# Ans 5:-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
```

```python
wine_data = pd.read_csv("winequality-red.csv")
```

```python
plt.figure(figsize=(12, 8))
for column in wine_data.columns:
    plt.subplot(3, 4, wine_data.columns.get_loc(column) + 1)
    sns.histplot(data=wine_data, x=column, kde=True)
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
for column in wine_data.columns:
    p_value = stats.shapiro(wine_data[column])[1]
    if p_value < 0.05:
        print(f"{column} does not follow a normal distribution (p-value: {p_value:.4f})")
```

```python
transformations = {
    "fixed acidity": "log",
    "residual sugar": "sqrt",
    "chlorides": "log",
    "free sulfur dioxide": "log",
    "total sulfur dioxide": "log",
    "density": "box-cox",
    "sulphates": "log",
    "alcohol": "log"
}
```

```python
for column, transformation in transformations.items():
    if column in wine_data.columns:
        if transformation == "log":
            wine_data[column] = np.log1p(wine_data[column])
        elif transformation == "sqrt":
            wine_data[column] = np.sqrt(wine_data[column])
        elif transformation == "box-cox":
            wine_data[column], _ = stats.boxcox(wine_data[column])
```

```python
plt.figure(figsize=(12, 8))
for column in wine_data.columns:
    plt.subplot(3, 4, wine_data.columns.get_loc(column) + 1)
    sns.histplot(data=wine_data, x=column, kde=True)
    plt.title(column)

plt.tight_layout()
plt.show()
```

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

```python
# Ans 6:-
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
wine_data = pd.read_csv("winequality-red.csv")
X = wine_data.drop(columns=['quality'])
y = wine_data['quality']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_

total_variance = 0
num_components = 0
for variance in explained_variance_ratio:
    total_variance += variance
    num_components += 1
    if total_variance >= 0.90:
        break

print("Minimum number of principal components to explain 90% variance:", num_components)
```
