```python
Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.
```

```python
# Ans :-1
import seaborn as sns
import plotly.express as px
titanic_data = sns.load_dataset("titanic")
fig = px.scatter(titanic_data, x="age", y="fare")
fig.show()
```

```python
Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
# Ans 2:-
import plotly.express as px
tips_data = px.data.tips()
fig = px.box(tips_data, x="day", y="total_bill", color="smoker", notched=True)
fig.show()
```

```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
```

```python
# Ans 3:-
import plotly.express as px
tips_data = px.data.tips()
fig = px.histogram(tips_data, x="sex", y="total_bill", color="day", pattern_shape="smoker")
fig.show()
```

```python
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
```

```python
# Ans 4:-
iris_data = px.data.iris()
fig = px.scatter_matrix(iris_data, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
# Ans 5:-
Distplot is a visualization tool used to display the distribution of a univariate set of observations
```

```python
import plotly.express as px
df = px.data.tips()
```

```python
fig = px.histogram(df, x="total_bill", nbins=30, opacity=0.7,
                   title="Distribution of Total Bill")
fig.add_trace(px.line(df, x="total_bill", y=df["total_bill"].plot.density()).data[0])
```
