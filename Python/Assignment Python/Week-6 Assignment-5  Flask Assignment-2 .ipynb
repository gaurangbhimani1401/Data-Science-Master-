Q1. Explain GET and POST methods.

```python
Ans 1:-
GET Method:
    The GET method is used to retrieve or fetch data from a specified resource.
    It sends data in the URL parameters as part of the query string.
    GET requests are idempotent, meaning multiple identical GET requests will produce the same result.
```

```python
POST Method:
    The POST method is used to submit or send data to be processed to a specified resource.
    It sends data in the body of the HTTP request.
    POST requests are not idempotent as multiple identical POST requests may have different effects on the server's state.
```

Q2. Why is request used in Flask?

```python
Ans 2:- The request object in Flask is used to access incoming request data and information.
        It provides a way to access the data sent by a client as part of an HTTP request.
```

```python
Retrieving Form Data:
    The request object allows you to access form data submitted by the client.
    It provides methods to access form inputs, checkboxes, radio buttons, and other form elements.
```

Q3. Why is redirect() used in Flask?

```python
Ans 3:- The redirect() function in Flask is used to redirect the user to a different URL or route within the application.
        It is primarily used for performing URL redirection, where the server instructs the clients browser to navigate to a different location.
```

```python
Route Navigation:
    If a user accesses a certain route or URL and you want to direct them to a different route or page, you can use redirect() to send them to the desired location.
    This is useful for managing route navigation within your application.
```

Q4. What are templates in Flask? Why is the render_template() function used?

```python
Ans 4:- In Flask, templates are used to separate the presentation logic from the application logic.
        A template is an HTML file with placeholders that can be dynamically filled with data from the Flask application.
        It allows you to create dynamic web pages by combining static HTML content with dynamic content generated by the application.
```
