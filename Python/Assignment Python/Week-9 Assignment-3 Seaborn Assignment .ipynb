Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Ans 1:-
Seaborn is a powerful Python library for data visualization.
It provides a high-level interface to create informative and visually appealing statistical graphics
```

```python
Line Plot:
    A line plot is useful for visualizing the relationship between two continuous variables over a continuous interval or time.
    It helps in understanding trends, patterns, and changes in the data.

Scatter Plot:
    A scatter plot is used to display the relationship between two continuous variables.
    It helps in identifying correlations, clusters, or outliers in the data.

Bar Plot:
    A bar plot is used to compare categorical variables by showing the values on one axis and the categories on the other axis.
    It helps in comparing data across different categories and identifying patterns or differences.

Histogram:
    A histogram is used to represent the distribution of a single variable by dividing it into bins and displaying the frequency or count
    of values within each bin.
    It helps in understanding the shape, skewness, and central tendency of the data.

Box Plot:
    A box plot, also known as a box-and-whisker plot, is used to display the distribution of a dataset and visualize summary 
    statistics such as quartiles, outliers, and the median.
    It helps in comparing distributions, identifying outliers, and understanding the spread of the data.
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
# Ans 2:-
import seaborn as sns
import matplotlib.pyplot as plt
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data
```

```python
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
sns.set(style="darkgrid")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Ans 3:-
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data
```

```python
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age by Passenger Class')
plt.show()
```

```python
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# Ans 4:-
import seaborn as sns
import matplotlib.pyplot as plt
diamonds_data = sns.load_dataset("diamonds")
```

```python
diamonds_data
```

```python
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices with Cut')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Ans 6:-
import seaborn as sns
flights_data = sns.load_dataset("flights")
```

```python
flights_data
```

```python
flights_matrix = flights_data.pivot("month", "year", "passengers")
sns.heatmap(data=flights_matrix, cmap="YlGnBu")
plt.show()
```
